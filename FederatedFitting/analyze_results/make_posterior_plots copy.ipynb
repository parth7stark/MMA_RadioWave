```python
import corner
import matplotlib.pyplot as plt

def make_1_corner_many(samples1, samples2, idx1, idx2, labels=params, true_values=None):
    """
    Overlay 2D corner plots for two sets of samples showing only the 90% contour (no shading).

    Parameters:
    - samples1, samples2: arrays of shape (n_samples, n_params)
    - idx1, idx2: indices of the two parameters to plot
    - labels: list of parameter names (optional)
    - true_values: list of true values for the parameters (optional)
    """

    # Slice the 2D subspace from each sample set
    subs1 = samples1[:, [idx1, idx2]]
    subs2 = samples2[:, [idx1, idx2]]
    
    # Labels and truths
    sub_labels = [labels[idx1], labels[idx2]] if labels else None
    sub_truths = [true_values[idx1], true_values[idx2]] if true_values is not None else None

    # First corner plot: no fill, 90% contour only
    fig = corner.corner(
        subs1,
        labels=sub_labels,
        show_titles=True,
        title_fmt=".2f",
        quantiles=[0.05, 0.5, 0.95],
        title_kwargs={"fontsize": 12},
        label_kwargs={"fontsize": 14},
        plot_datapoints=False,
        fill_contours=False,
        levels=(0.90,),
        smooth=1.0,
        smooth1d=1.0,
        color='C0',
        truths=sub_truths,
        contourf_kwargs={'alpha': 0},  #  suppress fill completely
    )

    # Second corner plot, overlaid
    corner.corner(
        subs2,
        fig=fig,
        labels=sub_labels,
        show_titles=False,
        plot_datapoints=False,
        fill_contours=False,
        levels=(0.90,),
        smooth=1.0,
        smooth1d=1.0,
        color='C1',
        contourf_kwargs={'alpha': 0},  #  again, suppress fill
    )

    plt.tight_layout()
    plt.savefig('./joint_posteriors.png')
    plt.show()
    return fig
```
